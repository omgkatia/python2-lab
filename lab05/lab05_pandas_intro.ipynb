Krok 1 — Series z listy i dict

```python
import pandas as pd
import numpy as np
```

```python
# Series z listy — domyślny indeks (0, 1, 2, ...)
oceny = pd.Series([4.5, 3.0, 5.0, 3.5, 4.0])
print(oceny)
```

Zadanie 1a: Utwórz Series pensje z dict:

```python
# Klucze: 'Anna', 'Jan', 'Ewa', 'Marek', 'Kasia'
# Wartości: 5500, 7200, 4800, 9100, 6300

pensje = pd.Series({'Anna': 5500, 'Jan': 7200, 'Ewa': 4800, 'Marek': 9100, 'Kasia': 6300})
print(pensje)
```

Zadanie 1b: Odpowiedz na pytania o pensje :

```python
# 1. Wyświetl imię osoby zarabiającej najwięcej (użyj idxmax())
# 2. Oblicz średnią pensję
# 3. Wyświetl imiona osób zarabiających powyżej średniej (filtrowanie boolean)
# 4. Policz ile osób zarabia poniżej 6000 (użyj len() lub .sum())

osoba = pensje.idxmax()
print(osoba)

srednia = pensje.mean()
print(srednia)

maska = pensje > srednia

powyzej_sredniej = pensje[maska]
print(powyzej_sredniej)

ponizej6000 = (pensje < 6000).sum()
print(ponizej6000)
```

Najwięcej zarabia Marek. Średnia pensja wszystkich pracowników wynosi 6580zł. Powyżej średniej zarabia Jan - 7200 i Marek - 9100, natomiast poniżej średniej zarabiają 2 osoby.

Krok 2 — Operacje na Series

Zadanie 1c: Oblicz pensje po zmianach:

```python
# 1. Oblicz pensje po podwyżce 10% dla wszystkich (pensje * 1.10)
# 2. Oblicz pensje po premii 500 zł dla wszystkich (pensje + 500)
# 3. Oblicz nową pensję = (pensja * 1.10) + 500
# 4. Wyświetl ranking nowych pensji od najwyższej (użyj sort_values)

podwyzka10 = pensje * 1.10
print(podwyzka10)

premia = pensje + 500
print(premia)

nowa_pensja = (pensje * 1.10) + 500
print(nowa_pensja)

nowa_pensja.sort_values(ascending=False)
```

```python
url = 'https://raw.githubusercontent.com/mwaskom/seaborn-data/master/penguins.csv'
penguins = pd.read_csv(url)
```

Zadania

Zadanie 3a: Pierwszy rzut oka

```python
# 1. Wyświetl head(10)
# 2. Wyświetl shape
# 3. Uruchom info() — ile brakujących wartości widzisz?

print(penguins.head(10))
print(penguins.shape)
penguins.info()
```

Dataset ma 21 brakujących wartości.

Zadanie 3b: Statystyki opisowe

```python
# 1. Uruchom describe() — co mówią ci średnie i odchylenia?
# 2. Uruchom describe(include='all') — co dodają kolumny tekstowe?
# 3. Jaka jest mediana masy ciała (body_mass_g)?

print(penguins.describe())

print(penguins.describe(include='all'))

print(penguins.body_mass_g)
```

Średnie wartości pokazują przeciętne cechy pingwinów w zbiorze danych - średnia długość dzioba to około 44mm, średnia masa ciała - 4201g, średnia głębokość dzioba - 17mm i średnia długość płetwy 200mm. Odchylenie standardowe pokazuję jak bardzo wartości różnią się od średniej. Dla długości dzioba jest to 5.45mm, dla głębokości dzioba prawie 2mm, długość płetwy - 14mm i w przypadku masy ciała jest to 801g.

Zadanie 3c: Rozkłady wartości

```python
# 1. Ile jest pingwinów każdego gatunku? (value_counts na 'species')
# 2. Ile pingwinów na każdej wyspie? (value_counts na 'island')
# 3. Rozkład płci (value_counts na 'sex') — czy jest zbalansowany?
# 4. Jaki procent to gatunek Adelie? (normalize=True)

print(penguins['species'].value_counts())

print(penguins['island'].value_counts())

print(penguins['sex'].value_counts())

print(penguins['species'].value_counts(normalize=True))

```

Najliczniejszym gatunkiem jest Adelie - 152, następnie Gentoo - 124 i Chinstrap - 68. Na wsypie Biscoe znajduję się 168 pingwinów, na wyspie Dream 124, natomiast na wyspie Torgersen jest ich najmniej - 52. Rozkład płci jest zbalansowany liczba samców i samic jest prawie taka sama (168 i 165). Pingwiny Adelie stanowią około 44%.

Zadanie 3d: Brakujące dane

```python
# 1. Policz brakujące wartości w każdej kolumnie (isna().sum())
# 2. Jaki procent danych brakuje w kolumnie 'sex'?
# Podpowiedź: braki / len(penguins) * 100

print(penguins.isna().sum())

braki = penguins['sex'].isna().sum()

procent = braki / len(penguins) * 100

print(procent)


```

Najwięcej braków jest w kolumnie sex - 11, a w pozostałych numerycznych kolumnach po 2 braki. W kolumnie 'sex' brakuje około 3% danych.

```python
tips = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv')
# Dodaj kolumnę z procentem napiwku
tips['tip_pct'] = (tips['tip'] / tips['total_bill'] * 100).round(1)
```

```python
# 1. Oblicz średni rachunek (total_bill)
# 2. Oblicz średni napiwek w procentach (tip_pct)
# 3. Wyświetl który dzień ma najwięcej rachunków (value_counts na 'day')
# 4. Policz ile rachunków jest od palaczy (value_counts na 'smoker')
# 5. Wyświetl describe() tylko dla kolumny 'tip' — jaka jest mediana napiwku?
# 6. Jaki procent rachunków pochodzi z obiadu (Dinner)?
# Podpowiedź: value_counts(normalize=True) na kolumnie 'time'
# 7. Ile unikalnych rozmiarów grup jest w danych? (nunique na 'size')

print(tips['total_bill'].mean())

print(tips['tip_pct'].mean())

print(tips['day'].value_counts())

print(tips['smoker'].value_counts())

print(tips['tip'].describe()) #mediana - 3

print(tips['time'].value_counts(normalize=True))

print(tips['size'].nunique()) #unikalne rozmiary grup - 6
```
