```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv')
print(f"Wczytano: {df.shape[0]} wierszy, {df.shape[1]} kolumn")
```

```python
df.info()
```

```python
df.describe()
```

Jaki był najwyższy rachunek?

```python
df['total_bill'].max()
```

Najwyższy rachunek był wysoki i wynosił 50.81.

	W którym dniu tygodnia jest największy ruch?

```python
df['day'].value_counts()
```

Największy ruch jest w sobotę.

Czy palący dają wyższe napiwki niż niepalący?

```python
df.groupby('smoker')['tip'].mean()
```

Nieznaczącą różnicą palący dają wyższee napiwki.

Jaki procent rachunku stanowi napiwek (średnio)?

```python
(df['tip'] / df['total_bill'] * 100).mean()
```

Średnio napiwek stanowi 16% rachunku.

```python
df.groupby('day')['tip'].mean().plot(kind='bar', title='Średni napiwek wg dnia')
plt.ylabel('Napiwek (USD)')
plt.tight_layout()
plt.show()
```

Niedziela przynosi największe napiwki. Dobrym rozwiązaniem byłoby zatrudnienie większej ilości kelnerów weekendowych.
